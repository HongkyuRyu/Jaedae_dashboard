```python
import pandas as pd
```

```python
df = pd.read_excel("data/9-10월_제대_전체.xlsx")

#TODO: 정렬Key(업체_품명기준)으로 정렬
df_sorted_업체 = df.sort_values(by='정렬Key(업체_품명기준)')

# TODO: 정렬Key(MES품명기준)으로 정렬
df_sorted_mes = df.sort_values(by='정렬Key(MES품명기준)')

df_sorted_업체.head()
```

```python
# 정렬Key(업체품명기준)으로 그룹화하여 Case분류

def classify_case(group):
    """업체_품명기준에 따른 Case 분류"""
    if (group["업체단가"] == group["표준단가"]).all():
        return "Case0"
    elif (group["업체단가"].nunique() == 1):
        return "Case1"
    else:
        return "Case2"

df_sorted_업체['업체_품명기준_Case'] = df_sorted_업체.groupby('정렬Key(업체_품명기준)', group_keys=False).apply(
    lambda group: pd.Series(classify_case(group), index=group.index)
)

df_sorted_mes['업체_품명기준_Case'] = df_sorted_mes.groupby('정렬Key(MES품명기준)', group_keys=False).apply(
    lambda group: pd.Series(classify_case(group), index=group.index)
)

# 결과 확인
df_sorted_업체.head()
```

```python
# Case0, Case1, Case2 데이터를 각각 필터링하여 출력
case0_data = df_sorted_업체[df_sorted_업체['업체_품명기준_Case'] == 'Case0']
case1_data = df_sorted_업체[df_sorted_업체['업체_품명기준_Case'] == 'Case1']
case2_data = df_sorted_업체[df_sorted_업체['업체_품명기준_Case'] == 'Case2']

## 확인완료
print(f"Case0 데이터 개수: {len(case0_data)}")
print(f"Case1 데이터 개수: {len(case1_data)}")
print(f"Case2 데이터 개수: {len(case2_data)}")
```
