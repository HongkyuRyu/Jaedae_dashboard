```python
import pandas as pd

df_merged = pd.read_excel("data/제대타입별_현황_2.xlsx")


df_merged['합계: 표준단가'] = df_merged.groupby('품명+너비+높이')['표준단가'].transform('sum')
df_merged['개수: 표준단가'] = df_merged.groupby('품명+너비+높이')['표준단가'].transform('count')

# '업체단가_MAX'에 대한 합계 및 개수 계산
df_merged['합계: 업체단가MAX'] = df_merged.groupby('품명+너비+높이')['업체단가_MAX'].transform('sum')
df_merged['개수: 업체단가MAX'] = df_merged.groupby('품명+너비+높이')['업체단가_MAX'].transform('count')

# 결과 확인
print(df_merged.head())
```
